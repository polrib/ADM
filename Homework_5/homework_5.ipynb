# [RQ2] Given a category as input we want to rank all of the nodes in V according to the following criteria:

The first category of the rank, always corresponds to the input category. 

The order of the remaining categories is given by:
$distance(C_0,C_i)=median(ShortestPath(C_0,C_i))$

In according to do that, we have to proceed in steps:

- We created the adjacency list

- We have implemented the bfs algorithm

- We created the block ranking and then, we calculated the median of the values


First of all, in order to speed up the calculations, let's filter the data link removing all the edges not belonging to the 
selected categories.

First, we create a set with all the ids we need.

```python
id_list = []

for articles_list in df_full['List of articles']:
    id_list = id_list + articles_list
    
id_list = list(set(id_list))

id_list
```

Now we can filter the data_link. We can exclude edges where 'start' and 'end' are not present in the id_list

```python
data_link_filtered = data_link[
    data_link['start'].isin(id_list)
    & data_link['end'].isin(id_list)
]

data_link_filtered
```

We created the adiacency list starting from the filtered data_link  

```python
# Group the table by the column start

data_link_grouped = data_link_filtered.groupby('start')

# Create the empy dictionary

adiacency_list = {}

# Iterate over the grouped data


for key,item in tqdm(data_link_grouped, total=len(data_link_grouped)):
    # If there are not outgoing edges from a node

    if data_link_grouped.get_group(key)["end"].tolist()[0] == 'None':
                   
        # Insert an empty set in the dictionary

        adiacency_list[key] = set()
    else:
        
         # Insert the destination nodes in the dictionary
        adiacency_list[key] = set(data_link_grouped.get_group(key)["end"].tolist())
    
adiacency_list
```

We can now calculate the distance of two categories at time.

Let's select 'English television actors" as C0.

```python
c0_articles = list(set(df_full['List of articles'][19]))
```

We decided to use _BFS(breadth-first search)_ that allowed us to find the minimun path from a source node.

(It's not been necessary use Dijkstra's algorithm because all the edges have the same weight). 

This method visits all the nodes level by level starting from an initial node (in our case, the starting nodes are the ones in c0_articles) that we can decide.






![title](files.jpg)

```python
from collections import deque 

def bfs(graph, start):
    
    #Put in queue the source node and we labeled it as visited.
    
    visited = {start}    
    queue = deque([(start, 0)])
    levels = {}

    
    while queue:
        
        #delete from the queue the node and examine it.
        
        current, path = queue.popleft()
        
        #We store into the dict levels the ending node as key and the distance from the starting node as value
        
        levels[current]=path
        
        try:
            for neighbor in graph.get(current,[]):
                
                #If the item sought was not in this node put in queue all the successors not yet visited of the node in analysis else the the search stops 
                
                if neighbor not in visited:
                    
                    visited.add(neighbor)
                    
                    queue.append((neighbor, path + 1))
        except KeyError:
            
            break
    
    return levels
```

We executed the BFS starting from each C0 article. When we got the BFS result, we iterated over it and, for each item, we checked to which category it belong.

Then we stored the distances of that item in a dictionary called block_ranking, under the key of its category.


```python
block_ranking={k:[] for k in df_full['Categories']}

for c0_article in tqdm(c0_articles, total=len(c0_articles)):
    dic_bfs= bfs(adiacency_list,c0_article)
    
    #we looped trough the categories in df_full
    
    for k in range(len(df_full)):
        
        #we looped trough the articles of the categories
        
        for node in df_full['List of articles'][k]:
            
            #if the node of the selected category is in dic_bfs we append it insidde the dictionary block_ranking
            
            if node in dic_bfs.keys(): 
                
                block_ranking[df_full['Categories'][k]].append(dic_bfs[node])

block_ranking
```

Now we can calculate the median for each category using the median function of numpy.

We, then insert inside the dictionary the name of the category with the respective median of the distances

```python
category_distances = {k:[] for k in df_full['Categories']}
    
for category, distances in tqdm(block_ranking.items(),total=len(block_ranking.items())):
    
    median_categories = pd.np.median(distances)
    category_distances[category] = median_categories
    
category_distances
```

We ordered from the closest to the furthest. Obviously, C0 is the closest.

```python
category_order = sorted( ((k,v) for v,k in category_distances.items()))
category_order
```

Now we had the order of the categories and so, we ordered the nodes.

First, we need to assign the nodes to a single category. Criteria is:

- If the article belongs to the input category (we will talk about this in RQ2) it belongs to that one.
- Otherwise, the category of the article will correspond, among the categories it belongs to, to the closest to the input category.

We can recreate the df_full

```python
# Initialize dataframe
df_full_ordered = pd.DataFrame(columns=['Categories','List of articles','Lenght'])

# Empty set. Here we'll store all the already used articles
used_articles = set()

# Loop over ordered categories
for median, category in category_order:

    # Extract all the articles of the category and remove the already used ones
    articles_list = list( set(list(df_full[df_full['Categories'] == category]['List of articles'])[0]) - used_articles )
    
    # Add the list to the used articles set
    for article in articles_list:
        used_articles.add(article)
    
    # Prepare data to append in the new dataframe
    data = pd.DataFrame(data={'Categories': [category], 'List of articles': ['0'],'Lenght': ['0']})
    data.at[0,'List of articles'] = articles_list
    data.at[0,'Lenght'] = len(articles_list)
    
    # Append data
    df_full_ordered = df_full_ordered.append(data)

# Reindex and print
df_full_ordered = df_full_ordered.reset_index(drop=True)
df_full_ordered
```

Now we have the categories ordered, and the articles with multiple categories associated to a single one. At this point we need only to rank the articles inside a category based on their score.

The score of an article is define as the number of inbound connections to that node.
We can get this information from the data_link structure.

Let's iterate on the df_full_ordered and than create a temporary structure in which we'll store, for each category, each node with its score. Then we order and append the ordered list of nodes to the final vector.

We can start creating a new adiacency list, but with edges going in the opposite direction

```python
data_link_grouped_end = data_link_filtered.groupby('end')

adiacency_list_inverted = {}

for key,item in tqdm(data_link_grouped_end, total=len(data_link_grouped_end)):
    if data_link_grouped_end.get_group(key)["start"].tolist()[0] == 'None':
        adiacency_list_inverted[key] = set()
    else:
        adiacency_list_inverted[key] = set(data_link_grouped_end.get_group(key)["start"].tolist())
    
adiacency_list_inverted
```

We created a dictionary that has as keys all the articles of all the categories.


```python
articles_score = {k:0 for k in id_list}
```

```python
# Empty set. Here we'll store all the already used articles
used_articles = set()

# We need to inizialize the score dict with the C0 values
# Score of the C0 nodes is the count of the incoming edges

row = df_full_ordered.iloc[0]
    
articles = row['List of articles']

for article in articles:
    used_articles.add(article)

for article in tqdm(articles, total=len(articles)):
    try:
        score = len( set(adiacency_list_inverted[article]).intersection(used_articles))
    except KeyError:
        score = 0
    articles_score[article] = score
    
articles_score
```

Now we can create the loop for the remaining categories

```python
# Loop over categories
for row in tqdm(df_full_ordered.iterrows(), total=34):
    
    # Skip C0
    if row[1]['Categories'] != 'Category:English_television_actors':
    
        articles = row[1]['List of articles']

        # Fill the used_articles set
        for article in articles:
            used_articles.add(article)

        # Loop over category articles
        for article in articles:
            
            # Initialize score
            score = 0
            
            try:
                # Check the adiacency list and select only articles in common with used_articles
                valid_articles = set(adiacency_list_inverted[article]).intersection(used_articles)
                
                # Sum scores of all the valid articles
                for valid_article in valid_articles:
                    score = score + articles_score[valid_article]
                    
                # Save score    
                articles_score[article] = score
                
            except KeyError:
                # Save score 
                articles_score[article] = score
    
articles_score
```

We created the final vector ordered by the categories and the scores

```python
import operator

block_ranking_final = []

# Loop over categories
for row in tqdm(df_full_ordered.iterrows(), total=35):
    articles = row[1]['List of articles']
    
    # Create empty dict for the category
    category_articles = {}
    
    # Fill the dict with the couple of values node and score
    for article in articles:
        category_articles[article] = articles_score[article]
    
    # Sort the dict
    sorted_cat = sorted(category_articles.items(), key=operator.itemgetter(1))
    
    # Fill the final vector
    for key, val in sorted_cat:
        block_ranking_final.append(key)
    
block_ranking_final
```
